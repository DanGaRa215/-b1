```python
# ファイルに保存したコピーされたデータのパス
file_path = r"C:\VScode\muds\zemi\-b1\txt\1.txt"


# 抽出したコメントを格納するリスト
extracted_comments = []

# ファイルを読み込む
with open(file_path, "r", encoding="utf-8") as f:
    lines = f.readlines()

# 4行1セットで処理（2行目を抽出）
for i in range(0, len(lines), 4):  # 4行ずつ処理する
    # 4行未満のデータをスキップ
    if i + 3 >= len(lines):  # データが4行未満の場合
        continue
    # 2行目（コメント部分）を抽出
    comment = lines[i + 1].strip()  # 余分な改行や空白を削除
    extracted_comments.append(comment)

# 抽出結果を表示
print("抽出されたコメント：")
for comment in extracted_comments:
    print(comment)

# 結果をCSVに保存
import pandas as pd
df = pd.DataFrame({"Comment": extracted_comments})
df.to_csv("extracted_comments.csv", index=False, encoding="utf-8")
print("抽出結果が 'extracted_comments.csv' に保存されました！")
```

```python
import re
import pandas as pd

# ファイルパス
file_path = r"C:\VScode\muds\zemi\-b1\txt\1.txt"

# 抽出したコメントを格納するリスト
extracted_comments = []

# ファイルを読み込む
with open(file_path, "r", encoding="utf-8") as f:
    lines = f.readlines()

# 正規表現を使用して、数字の次の行を抽出
for i in range(len(lines) - 1):
    # 数字を含む行を判定（例: "34週間前"）
    if re.search(r"\d+週間前", lines[i]):
        comment = lines[i + 1].strip()  # 次の行を抽出
        if comment:  # コメントが空でない場合のみ追加
            extracted_comments.append(comment)

# 抽出結果を確認
print("抽出されたコメント：")
print(extracted_comments)

# 結果をCSVに保存
df = pd.DataFrame({"Comment": extracted_comments})
df.to_csv("aespa1.csv", index=False, encoding="utf-8")
print("結果が 'aespa1.csv' に保存されました。")
```
