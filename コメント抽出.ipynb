```python
import pandas as pd
from langdetect import detect
import logging

# ãƒ­ã‚°ã®è¨­å®š
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# CSVãƒ‡ãƒ¼ã‚¿ã®èª­ã¿è¾¼ã¿
data = pd.read_csv("CSV/all_comments.csv")
logging.info("CSV file loaded successfully.")

# è‹±èªã‚³ãƒ¡ãƒ³ãƒˆã‚’ãƒ•ã‚£ãƒ«ã‚¿ãƒªãƒ³ã‚°ã™ã‚‹é–¢æ•°
def is_english(comment):
    try:
        lang = detect(comment)
        logging.debug(f"Detected language: {lang} for comment: {comment[:50]}")  # æœ€åˆã®50æ–‡å­—ã‚’è¡¨ç¤º
        return lang == 'en'
    except Exception as e:
        logging.warning(f"Language detection failed for comment: {comment[:50]} - {str(e)}")
        return False

# å„ã‚³ãƒ¡ãƒ³ãƒˆãŒè‹±èªã‹ã©ã†ã‹ã‚’åˆ¤å®š
logging.info("Starting language detection...")
data['Is English'] = data['Comment'].apply(is_english)
logging.info("Language detection completed.")

# è‹±èªã®ã‚³ãƒ¡ãƒ³ãƒˆã®ã¿æŠ½å‡º
english_comments = data[data['Is English']].drop(columns=['Is English'])
logging.info(f"Filtered {len(english_comments)} English comments out of {len(data)} total comments.")

# çµæœã‚’æ–°ã—ã„CSVã«ä¿å­˜
output_file = "english_comments_with_logs.csv"
english_comments.to_csv(output_file, index=False, encoding="utf-8")
logging.info(f"Filtered English comments saved to {output_file}.")
```

```python
import pandas as pd
from langdetect import detect
from deep_translator import GoogleTranslator

# CSVãƒ•ã‚¡ã‚¤ãƒ«ã®èª­ã¿è¾¼ã¿
file_path = 'CSV/all_comments.csv'  # ã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰ã•ã‚ŒãŸãƒ•ã‚¡ã‚¤ãƒ«ã®ãƒ‘ã‚¹
data = pd.read_csv(file_path)

# ãƒ­ã‚°ç”¨ãƒªã‚¹ãƒˆã‚’ä½œæˆ
logs = []

# ç¿»è¨³é–¢æ•°ã®å®šç¾©
def translate_to_english(comment, index):
    try:
        detected_language = detect(comment)  # è¨€èªã‚’æ¤œå‡º
        if detected_language != 'en':  # è‹±èªä»¥å¤–ã®ã‚³ãƒ¡ãƒ³ãƒˆã‚’ç¿»è¨³
            translated = GoogleTranslator(source='auto', target='en').translate(comment)
            logs.append(f"Row {index}: Translated from {detected_language} to English.")
            return translated
        else:
            logs.append(f"Row {index}: Already in English.")
            return comment
    except Exception as e:
        logs.append(f"Row {index}: Error - {e}")
        return comment  # ç¿»è¨³ã«å¤±æ•—ã—ãŸå ´åˆã¯å…ƒã®ã‚³ãƒ¡ãƒ³ãƒˆã‚’è¿”ã™

# ã‚³ãƒ¡ãƒ³ãƒˆåˆ—ã‚’ç¿»è¨³
data['Translated_Comment'] = [
    translate_to_english(comment, idx) for idx, comment in enumerate(data['Comment'].astype(str))
]

# ç¿»è¨³çµæœã‚’æ–°ã—ã„CSVãƒ•ã‚¡ã‚¤ãƒ«ã¨ã—ã¦ä¿å­˜
output_file_path = 'CSV'
data.to_csv(output_file_path, index=False, encoding='utf-8-sig')

# ãƒ­ã‚°ã‚’è¡¨ç¤º
log_file_path = 'txt/translation_logs.txt'
with open(log_file_path, 'w', encoding='utf-8') as log_file:
    log_file.write("\n".join(logs))

print(f"ç¿»è¨³çµæœã‚’ {output_file_path} ã«ä¿å­˜ã—ã¾ã—ãŸï¼")
print(f"ãƒ­ã‚°ã‚’ {log_file_path} ã«ä¿å­˜ã—ã¾ã—ãŸï¼")
```
